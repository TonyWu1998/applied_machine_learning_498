```python
# for the 2.4
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import ShuffleSplit
%matplotlib inline


traindata = pd.read_csv('adult.data', header = None)
testdata = pd.read_csv('adult.test', header = None)


# a helper function to display labels as values
def displayLabel(row):
    if row[14] == " <=50K":
        return -1
    else:
        return 1


def get_accu(a, b, val_features, val_labels, size):
    prediction = val_features.dot(a) + b
    result = np.where(prediction > 0, 1, -1).reshape(size,)
    accuracy = np.sum(result == val_labels) / val_labels.shape[0]
    return accuracy

# dropping the non-contineous colmns.
traindata = traindata.dropna(axis = 0, how = 'any')
labels = traindata.apply(lambda row: displayLabel(row), axis=1).values

traindata = traindata.drop(columns=[1,3,5,6,7,8,9,13])
features = traindata.drop(columns=14).values
features = features.astype(float)

#normalize the data
for x in range(features.shape[1]):
    temp = features[:, x]
    features[:, x] = (temp - temp.mean() / temp.std())


# helper fucntion for the gradient descent
def gradientdescent(X, Y, a, b, lamb):
    matrix = Y * (np.dot(X, a) + b)
    bigger = np.where(matrix >= 1)
    smaller = np.where(matrix < 1)
    lengthBig = len(bigger[0])*lamb*a
    lengthSmall = len(smaller[0])*lamb*a-Y[smaller].T.np.dot(X[smaller])
    return (lengthBig + lengthSmall, -np.sum(Y[smaller]))


def SVM(labels, features, lamb):
    a = np.random.random([6,1])
    b = np.random.random()

    accur = []
    magni = []

    for s in range(1,51):
        step_length = 1/(0.01 * s + 50)

        mask = np.array(range(0, len(labels)))
        np.random.shuffle(mask)
        newfeatures = np.array(features[mask[-50:]])
        newlabels = np.array(labels[mask[-50:]])
        # here to split the portion of the test data.
        dataArray = np.array_split(mask[:-50], STEPS)

        for each in range(300):
            batchsize = len(dataArray[each])
            np.shape(features)
            X = features[dataArray[each]]
            print(dataArray[each])
            print(len(dataArray[each]))
            Y = labels[dataArray[each]].reshape(batchsize, 1)
            Ga, Gb = gradientdescent(X, Y, a, b, lamb)
            a = a - step_length * (1/batchsize)*Ga
            b = b - step_length * (1/batchsize)*Gb

            if (each % 30 == 0):
                accur.append(get_accu(a, b, newfeatures, newlabels, 50))
                magni.append(np.linalg.norm(a))
    return (a, b, accur, magni)


STEPS = 300
lambvalues = np.array([1e-3,0.0015,0.002,0.0025,0.003,1e-2,1e-1,1])

trainFeatures, trainLabels, testFeatures, testLabel = train_test_split(features, labels, test_size = 0.1)


accu_all = []
avg_accu = []
mag_all = []
steps = np.arange(500)

for reg in lambvalues:
    a, b, accu, mag = SVM(trainLabels, trainFeatures, reg)
    accu_all.append(accu)
    mag_all.append(mag)

for accu in accu_all:
    avg_accu.append(sum(accu)/len(accu))

print(avg_accu)









    
```
