```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt

df = pd.read_csv('iris.data',header = None)
print(df)
df = df.iloc[:, :-1]
print(df)

mu, sigma1, sigma2, sigma3, sigma4 = 0, 0.1, 0.2, 0.5, 1.0 # mean and standard deviation
s1 = np.random.normal(mu, sigma1, 150*4)
s1 = s1.reshape(150,4)
pd.DataFrame(s1)
df_noise1 = df + s1

s2 = np.random.normal(mu, sigma2, 150*4)
s2 = s2.reshape(150,4)
pd.DataFrame(s2)
df_noise2 = df + s2

s3 = np.random.normal(mu, sigma3, 150*4)
s3 = s3.reshape(150,4)
pd.DataFrame(s3)
df_noise3 = df + s3

s4 = np.random.normal(mu, sigma4, 150*4)
s4 = s4.reshape(150,4)
pd.DataFrame(s4)
df_noise4 = df + s4

pca = PCA(n_components = 1)
pca1 = PCA(n_components = 2)
pca2 = PCA(n_components = 3)
pca3 = PCA(n_components = 4)

# pca = 1

principalComponents1a = pca.fit_transform(df_noise1)
X_ori1a = pca.inverse_transform(principalComponents1a)
square1a = np.square(df_noise1 - X_ori1a)
t1 = np.sum(np.sum(square1a))/(150*4)

principalComponents1b = pca.fit_transform(df_noise2)
X_ori1b = pca.inverse_transform(principalComponents1b)
square1b = np.square(df_noise1 - X_ori1b)
t2 = np.sum(np.sum(square1b))/(150*4)

principalComponents1c = pca.fit_transform(df_noise3)
X_ori1c = pca.inverse_transform(principalComponents1c)
square1c = np.square(df_noise1 - X_ori1c)
t3 = np.sum(np.sum(square1c))/(150*4)

principalComponents1d = pca.fit_transform(df_noise4)
X_ori1d = pca.inverse_transform(principalComponents1d)
square1d = np.square(df_noise1 - X_ori1d)
t4 = np.sum(np.sum(square1d))/(150*4)

mse = [t1, t2, t3, t4]

# PCA = 2
principalComponents1a = pca1.fit_transform(df_noise1)
X_ori1a = pca1.inverse_transform(principalComponents1a)
square1a = np.square(df_noise1 - X_ori1a)
t1 = np.sum(np.sum(square1a))/(150*4)

principalComponents1b = pca1.fit_transform(df_noise2)
X_ori1b = pca1.inverse_transform(principalComponents1b)
square1b = np.square(df_noise1 - X_ori1b)
t2 = np.sum(np.sum(square1b))/(150*4)

principalComponents1c = pca1.fit_transform(df_noise3)
X_ori1c = pca1.inverse_transform(principalComponents1c)
square1c = np.square(df_noise1 - X_ori1c)
t3 = np.sum(np.sum(square1c))/(150*4)

principalComponents1d = pca1.fit_transform(df_noise4)
X_ori1d = pca1.inverse_transform(principalComponents1d)
square1d = np.square(df_noise1 - X_ori1d)
t4 = np.sum(np.sum(square1d))/(150*4)

mse1 = [t1, t2, t3, t4]

# PCA = 3
principalComponents1a = pca2.fit_transform(df_noise1)
X_ori1a = pca2.inverse_transform(principalComponents1a)
square1a = np.square(df_noise1 - X_ori1a)
t1 = np.sum(np.sum(square1a))/(150*4)

principalComponents1b = pca2.fit_transform(df_noise2)
X_ori1b = pca2.inverse_transform(principalComponents1b)
square1b = np.square(df_noise1 - X_ori1b)
t2 = np.sum(np.sum(square1b))/(150*4)

principalComponents1c = pca2.fit_transform(df_noise3)
X_ori1c = pca2.inverse_transform(principalComponents1c)
square1c = np.square(df_noise1 - X_ori1c)
t3 = np.sum(np.sum(square1c))/(150*4)

principalComponents1d = pca2.fit_transform(df_noise4)
X_ori1d = pca2.inverse_transform(principalComponents1d)
square1d = np.square(df_noise1 - X_ori1d)
t4 = np.sum(np.sum(square1d))/(150*4)

mse2 = [t1, t2, t3, t4]

# PCA = 4
principalComponents1a = pca3.fit_transform(df_noise1)
X_ori1a = pca3.inverse_transform(principalComponents1a)
square1a = np.square(df_noise1 - X_ori1a)
t1 = np.sum(np.sum(square1a))/(150*4)

principalComponents1b = pca3.fit_transform(df_noise2)
X_ori1b = pca3.inverse_transform(principalComponents1b)
square1b = np.square(df_noise1 - X_ori1b)
t2 = np.sum(np.sum(square1b))/(150*4)

principalComponents1c = pca3.fit_transform(df_noise3)
X_ori1c = pca3.inverse_transform(principalComponents1c)
square1c = np.square(df_noise1 - X_ori1c)
t3 = np.sum(np.sum(square1c))/(150*4)

principalComponents1d = pca3.fit_transform(df_noise4)
X_ori1d = pca3.inverse_transform(principalComponents1d)
square1d = np.square(df_noise1 - X_ori1d)
t4 = np.sum(np.sum(square1d))/(150*4)

mse3 = [t1, t2, t3, t4]

mse_total = [mse, mse1, mse2, mse3]
x = [0.1, 0.2, 0.5, 1.0]
plt.figure()
plt.plot(x, mse, marker = 'o', label = 'Principal components 1')
plt.plot(x, mse1,marker = 'o', label = 'Principal components 2')
plt.plot(x, mse2,marker = 'o', label = 'Principal components 3')
plt.plot(x, mse3,marker = 'o', label = 'Principal components 4')
plt.legend()
plt.draw()

```
